### Data cleaning end and (university dataset start)

```python
university_towns = []
with open(r"C:\Users\admin\Downloads\university_towns.txt") as file:
     for line in file:
         if '[edit]' in line:
             state = line
         else:
             university_towns.append((state, line))
             
university_towns[:5]
```

```python
towns_df = pd.DataFrame(university_towns,columns=['State' , 'RegionName'])
```

```python
towns_df.head()
```

```python
def get_citystate(item):
    if ' (' in item:
        return item[:item.find(' (')]
    elif '[' in item:
        return item[:item.find('[')]
    else:
        return item
```

```python
towns_df = towns_df.map(get_citystate)
```
