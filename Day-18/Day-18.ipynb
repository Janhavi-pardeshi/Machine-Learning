Convert to a Classification Task


```python
clean_df = df.copy()
clean_df['high_humidity_label'] = (clean_df['relative_humidity_3pm'] > 24.99)*1
print(clean_df['high_humidity_label'])
```

Target is stored in 'y'.


```python
y=clean_df[['high_humidity_label']].copy()
```

```python
y.head()
```

```python
clean_df['relative_humidity_3pm'].head()
y.head()
```

use 9am sensor signals as features humidity at 3pm

```python
morning_features = ['air_pressure_9am','air_temp_9am','avg_wind_direction_9am','avg_wind_speed_9am',
        'max_wind_direction_9am','max_wind_speed_9am','rain_accumulation_9am',
        'rain_duration_9am']
```

```python
X = clean_df[morning_features].copy()
```

```python
X.columns
```

```python
y.columns
```

perform test and train split

```python
X_train,X_test,y_train,y_test  = train_test_split(X,y, test_size=0.33,random_state=0)
```

fit on train

```python
humidity_classifier = DecisionTreeClassifier(max_leaf_nodes=10, random_state=0)
humidity_classifier.fit(X_train, y_train)
```

```python
type(humidity_classifier)
```

Predict in Test Set

```python
predictions = humidity_classifier.predict(X_test)
```

```python
predictions[:10]
```

```python
y_test['high_humidity_label'][:10]
```

Measure accuracy of the classifier

```python
accuracy_score(y_true = y_test, y_pred = predictions)
```

## Using try..Except block

```python
print ("Practicing for try block")
try:
    numerator=50
    denom=int(input("Enter the denominator"))
    quotient=(numerator/denom)
    print(quotient)
    print ("Division performed successfully")
except ZeroDivisionError:
    print ("Denominator as ZERO.... not allowed")
print("OUTSIDE try..except block")
```

## Multiple Exception

```python
print ("Handling multiple exceptions")
try:
    numerator=50
    denom=int(input("Enter the denominator: "))
    print (numerator/denom)
    print ("Division performed successfully")
except ZeroDivisionError:
    print ("Denominator as ZERO is not allowed")
except ValueError:
    print("Only INTEGERS should be entered")
```

```python
print ("Handling exceptions without naming them")
try:
    numerator=50
    denom=int(input("Enter the denominator"))
    quotient=(numerator/denom)
    print ("Division performed successfully")
except ValueError:
    print ("Only INTEGERS should be entered")
except:
    print(" OOPS.....SOME EXCEPTION RAISED")
```

## Use of else clause

```python
print ("Handling exception using try...except...else")
try:
    numerator=50
    denom=int(input("Enter the denominator: "))
    quotient=(numerator/denom)
    print ("Division performed successfully")
except ZeroDivisionError:
    print ("Denominator as ZERO is not allowed")
except ValueError:
    print ("Only INTEGERS should be entered")
else:
    print ("The result of division operation is ", quotient)
```

## Finally clause

```python
print ("Handling exception using try...except...else...finally")
try:
    numerator=50
    denom=int(input("Enter the denominator: "))
    quotient=(numerator/denom)
    print ("Division performed successfully")
except ZeroDivisionError:
    print ("Denominator as ZERO is not allowed")
except ValueError:
    print ("Only INTEGERS should be entered")
else:
    print ("The result of division operation is ", quotient)
finally:
    print ("OVER AND OUT")
```

```python
# Write a program to find if the product of consequective number in a list is x
# x = 210
# find:(5*6*7 == x)
```

```python
list = [-4,5,6,-2,7,120]
x = 210

def consecutive_product(list, x):
    for i in range(len(list) - 2):
        product = list[i] * list[i + 1] * list[i + 2]
        if product == x:
            return True
    return False

if consecutive_product(list, x):
    print(f"The product of three consecutive numbers in the list equals to {x}.")
else:
    print(f"No three consecutive numbers in the list have a product that equals to {x}.")

```
