```python
df['Group_nums'] = df.PassengerId.apply(lambda x: x.split('_')).apply(lambda x: x[0])
df.PassengerId
```

```python
df['Grouped'] = ((df['Group_nums'].value_counts() > 1).reindex(df['Group_nums'])).tolist()
df.Grouped
```

```python
df['Deck'] = df.Cabin.apply(lambda x: str(x).split('/')).apply(lambda x: x[0])
df.Deck
```

```python
df['Side'] = df.Cabin.apply(lambda x: str(x).split('/')).apply(lambda x: x[2])
df.Side
```

```python
df['Has_expenses'] = df['Expenses'] > 0
df.Has_expenses
```
